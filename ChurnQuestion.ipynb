```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

df_original = pd.read_csv('BankChurners.csv')
dropColumns = ['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 
               'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']
df = df_original.drop(columns=dropColumns)
df.head()
```

```python
df['Dependent_count'] = df['Dependent_count'].astype(str)
df_dummies = pd.get_dummies(df, drop_first=True)
df_dummies.head()
```

```python
df_dummies = df_dummies.drop(['CLIENTNUM'], axis=1)
```

```python
columns = df_dummies.columns.values
columns
```

```python
index = np.where(columns == 'Attrition_Flag_Existing Customer')[0]
X_indexes = list(np.delete(columns, index))
y_indexes = ['Attrition_Flag_Existing Customer']

X = df_dummies[X_indexes]
y = df_dummies[y_indexes]

X_standarized = StandardScaler().fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_standarized, y, test_size = .30, random_state=42)

clf = LogisticRegression(random_state = 0, solver='lbfgs', max_iter=200).fit(X_train, y_train)
y_pred = clf.predict(X_test)
clf.score(X_test, y_test)
```

```python
cf = confusion_matrix(y_test, y_pred)
cf
```

This model has an accuracy of 89.83%. In the confusion matrix we can see the majority of the errors are type 1 and the errors type 2 are quite low. This model is pretty decent, can be improved but this is not the current goal. We need to check the coefficients of the model in order to see which variables are more realated to the churn action.

```python
coef = pd.DataFrame(clf.coef_[0], index = X_indexes, columns = ['Coefficients'])
coef
```

```python
no_dummie_coef = coef.iloc[0:13, 0]
dummie_coef = coef.iloc[13:len(coef), 0]
```

```python
dummie_coef.sort_values()
```

It looks like the most important feature from the dummies variables is the Gender. We can see that being male impacts positively on Existing client which means that the females are more prone to churn the bank, this is something that we already measured. We also can see that the people with lower income are more prone to stay in the bankin the same way people with status Married, and people who don't declare the income.

```python
no_dummie_coef.sort_values()
```

This variables are a bit more complicate to interpret due to the difference between measures units.
